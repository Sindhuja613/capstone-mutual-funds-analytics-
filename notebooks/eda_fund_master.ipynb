```python
import pandas as pd

# =====================================================================
# STEP 1: LOAD THE TWO EXPLICIT DATASETS
# (Ensure your relative paths and filenames match your local folder setup)
# =====================================================================
df_master = pd.read_csv("../data/raw/01_fund_master.csv")
df_history = pd.read_csv("../data/raw/02_nav_history.csv")

# =====================================================================
# STEP 2: CONVERT CODE COLUMNS TO SETS FOR COMPREHENSIVE COMPRESSION
# =====================================================================
master_codes = set(df_master['amfi_code'].unique())
history_codes = set(df_history['amfi_code'].unique())

# Calculate intersections and differences
missing_in_history = master_codes - history_codes
orphaned_in_master = history_codes - master_codes

total_master_schemes = len(master_codes)
total_history_schemes = len(history_codes)
missing_count = len(missing_in_history)
orphan_count = len(orphaned_in_master)
```

```python
# STEP 3: CONSTRUCT THE REPORT TEXT
report_lines = []
report_lines.append("=" * 65)
report_lines.append("DATA QUALITY & INTEGRITY REPORT: SCHEME CODE VALIDATION")
report_lines.append("=" * 65)
report_lines.append(f"🔹 Total Unique Schemes cataloged in 'fund_master' : {total_master_schemes}")
report_lines.append(f"🔹 Total Unique Schemes recorded in 'nav_history'  : {total_history_schemes}")
report_lines.append("-" * 65)

report_lines.append("METRIC 1: REFERENTIAL INTEGRITY (Master -> History)")
if missing_count == 0:
    report_lines.append(" STATUS: PASSED")
    report_lines.append("   All schemes defined in the master data exist in the transaction logs.")
else:
    report_lines.append("⚠️ STATUS: WARNING / GAP DETECTED")
    report_lines.append(f"   {missing_count} scheme(s) in 'fund_master' are completely missing from 'nav_history'.")
    report_lines.append(f"   Missing Codes: {list(missing_in_history)}")

report_lines.append("-" * 65)
report_lines.append(" METRIC 2: ORPHAN RECORDS CHECK (History -> Master)")
if orphan_count == 0:
    report_lines.append("STATUS: CLEAN")
    report_lines.append("   No untracked historical records found. All history links back to a valid master.")
else:
    report_lines.append("⚠️ STATUS: UNTRACKED CODES FOUND")
    report_lines.append(f"   {orphan_count} scheme(s) exist in 'nav_history' but are missing from 'fund_master'.")
    report_lines.append(f"   Orphaned Codes: {list(orphaned_in_master)}")

report_lines.append("=" * 65)
report_lines.append("End of Validation Report")
report_lines.append("=" * 65)

# Join lines together into a single string block
full_report_text = "\n".join(report_lines)
```

```python
# STEP 4: OUTPUT TO JUPYTER CELL AND SAVE TO FILE
print(full_report_text)

# Setup path to go up one directory out of 'notebooks/' and into 'reports/'
reports_dir = os.path.join("..", "reports")

report_file_path = os.path.join(reports_dir, "data_validation_report.txt")

# Write the file down locally
with open(report_file_path, "w", encoding="utf-8") as file:
    file.write(full_report_text)

print(f"\n📂 Executive text report successfully saved to: {report_file_path}")
```
